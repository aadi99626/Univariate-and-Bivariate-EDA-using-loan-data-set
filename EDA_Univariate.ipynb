<h1><center>EXPLORATORY DATA ANALYSIS</center></h1>

In this document we are going to do Exploratory data analysis for a data set. EDA is nothing but describing, analyzing, visualizing a data set to extract the insights from it.
The data set we are going to analyze, seems to be related to banking loan services.
In this data set we have used different python functions and charts to describe and visualize the data.


Charts: 

1. Histogram -  This chart is used to determine the frequency distribution for a variable. It is represented by rectangles called 'Bins'. Width of bins is the size of class interval.

2. Bar chart - This chart is used for the representation of categorical variables. The height of bar is equal to the value of the particular categorical variable.

3. Area chart - This chart is a good in visualizing a volume change over a period of time. It is based on the line chart. It gives a sense of summation of the quantitative data.

4. Line chart - This chart gives a visual representation of the trend of a variable in a certain period of time.

5. Box plot - This chart is useful to depict the groups of data based on quartiles. The lower box line represents Quartile 1 and the upper box line represent Quartile 3. Q3-Q1 which is the area between the upper and lower box line is called Inter Quartile range. With box plot it is easy to identify the outliers.

6. Ogive - This chart is line chart of Cumulative frequency less than and Cumulative frequency more than type. The intersection of these two lines gives the median.

<h2>Box Plot for Amount</h2>

This is the representation of Box plot chart for Amount data.
The black spots on the box plot represents the data points.
Data Description :  Minimum = 1000, Maximum = 7500, Quartile 1 (Q1) = 1000,Quartile 2 (Q2) or Median= 1600, Quartile 3 (Q3)= 2537.5.

Inter Quartile Range = Quartile 3 (Q3) - Quartile 1(Q1) = 2537.5 - 1000 = 1537.5
IQR is the range of data points enclosed within Q3 and  Q1.

We can see that the left whisker lies on the Q1, that is because the Minimum value and Q1 is same and this both are starting from same place.
Also we can see the some data points are outsidet the right whisker. As the data points are outside Min-Max range, we should consult the subject matter expert to verify if those data points are to be considered as outliers. 

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
ax = sns.boxplot(x=data_set["Amount"],color='tab:orange')
ax = sns.swarmplot(x=data_set["Amount"], data=data_set, color="0.25")
```
