<H2>OGIVE GRAPH</H2>

```python
data_set123= data_set['Age']
data_set123 = pd.DataFrame(data_set123)
data_set123 = pd.DataFrame(data_set.groupby(['Age'])['Age'].count())
data_set123['Age1'] = data_set123.index
data_set123 = data_set123.rename(columns={ "Age1": "Age","Age": "Age_Frequency"})
data_set123 = data_set123.reset_index(drop = True)
data_set123 = data_set123[['Age','Age_Frequency']]
import numpy as np
data_set123['CF-Less than'] = data_set123['Age_Frequency'].cumsum()
```

```python
d = {'CF-More than': [50,42,41,34,31,26,19,14,6,4]}
df = pd.DataFrame(data=d)
result = pd.concat([data_set123, df], axis=1)
result
```

```python

# importing package
import matplotlib.pyplot as plt
import numpy as np
  
# create data
x = data_set123['Age']
y = data_set123['CF-Less than']
z = result['CF-More than']
plt.figure(figsize= (10,8)) 
# plot lines
plt.plot(x, y, color='brown', marker='o', linestyle='-',label = "line 1")
plt.plot(x, z, color='blue', marker='o', linestyle='-',label = "line 1")

plt.legend()
plt.show()
```
